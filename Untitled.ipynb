# Carregamento dos dados

```python
from pandas import read_csv

url = "./SA_heart.csv"
dataset = read_csv(url, header=0, usecols=[i+1 for i in range(10)])

dataset.head(5)
```

# Preparação dos dados

### Extração de entradas e saídas, conversão da dataframe para numpy array


```python
import numpy as np

# extração da variável a ser predita
y = np.array(dataset['chd'])
dataset = dataset.drop('chd', axis=1)

# armazenamento do nome de cada variável preditora para uso futuro
X_names = list(dataset.columns)

# conversão para numpy array
X = np.array(dataset)
```

# Análise estatística inicial dos dados

### Quantidades médias

```python
from matplotlib import pyplot

pyplot.figure(figsize=(17,5))
pyplot.bar([i+1 for i in range(9)], [f.mean() for f in X.transpose()], tick_label=X_names, width=0.8)
pyplot.xlabel('Variáveis')
pyplot.ylabel('Média')
pyplot.show()
```

### Desvio padrão

```python
pyplot.figure(figsize=(17,5))
pyplot.bar([i+1 for i in range(9)], [f.std() for f in X.transpose()], tick_label=X_names, width=0.8)
pyplot.xlabel('Variáveis')
pyplot.ylabel('Desvio padrão')
pyplot.show()
```

# Criação dos modelos

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# instanciação dos modelos
models = []
models.append(('Árvore de decisão ID3',DecisionTreeClassifier(random_state=1, criterion='entropy')))
models.append(('Floresta randômica todas',RandomForestClassifier(random_state=1, n_estimators=100, max_features=9)))
models.append(('Floresta randômica sqrt',RandomForestClassifier(random_state=1, n_estimators=100, max_features=3)))
```

# Validação cruzada

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import RocCurveDisplay
from sklearn.metrics import auc

# Trecho do código responsável por plotar a curva ROC foi retirado de:
# https://scikit-learn.org/stable/auto_examples/model_selection/plot_roc_crossval.html

tprs = []
aucs = []
mean_fpr = np.linspace(0, 1, 100)

# iterar entre todos os modelos
for m in models:
    # validação cruzada em 10 rodadas com 10% dos dados separados para teste
    cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=1)
    

    fig, ax = pyplot.subplots(figsize=(12.4,9.6))
    
    # iteração entre todas as "fold's" do conjunto de dados
    for i, (train, test) in enumerate(cv.split(X, y)):
        
        # treinamento do modelo 
        m[1].fit(X[train], y[train])
        display = RocCurveDisplay.from_estimator(m[1], X[test], y[test], name='ROC fold {}'.format(i), ax=ax)
        
        interp_tpr = np.interp(mean_fpr, display.fpr, display.tpr)
        interp_tpr[0] = 0.0
        tprs.append(interp_tpr)
        aucs.append(display.roc_auc)
    
    # curva-base  
    ax.plot([0, 1], [0, 1], linestyle="--", lw=2, color="r", label="Chance", alpha=0.8)
    
    mean_tpr = np.mean(tprs, axis=0)
    mean_tpr[-1] = 1.0
    mean_auc = auc(mean_fpr, mean_tpr)
    std_auc = np.std(aucs)
    ax.plot(mean_fpr, mean_tpr, color='b',
        label=r'Mean ROC (AUC = %0.2f $\pm$ %0.2f)' % (mean_auc, std_auc),
        lw=2, alpha=.8)

    std_tpr = np.std(tprs, axis=0)
    tprs_upper = np.minimum(mean_tpr + std_tpr, 1)
    tprs_lower = np.maximum(mean_tpr - std_tpr, 0)
    ax.fill_between(mean_fpr, tprs_lower, tprs_upper, color='grey', alpha=.2,
                label=r'$\pm$ 1 std. dev.')

    ax.set(xlim=[-0.05, 1.05], ylim=[-0.05, 1.05],
       title=m[0])
    ax.legend(loc="lower right") 
    pyplot.show()
    
    
                                                              
```
